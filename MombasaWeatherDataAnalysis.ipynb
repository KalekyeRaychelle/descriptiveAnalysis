### FIND HOT DAYS

```python
#days with temp greater that thirty
hotDays=dataset[dataset["Temperature (°C)"]>30]
```

```python
print(hotDays)
```

### FIND COLD DAYS

```python
coldDays=dataset[dataset["Temperature (°C)"]<20]
print(coldDays)
```

### MEAN AMOUNT OF RAINFALL

```python
mean_rainfall=dataset["Rainfall (mm)"].mean()
print( "Average rainfall in Mombasa is: ",mean_rainfall," mm")
```

### TOTAL, MAXIMUM AND MINIMUM AMOUNT OF RAINFALL

```python
max_rainfall=dataset["Rainfall (mm)"].max()
min_rainfall=dataset["Rainfall (mm)"].min()
total_rainfall=dataset["Rainfall (mm)"].sum()

print("Maximum Rainfall:",max_rainfall,"mm")
print("Minimum Rainfall:",min_rainfall,"mm")
print("Total Rainfall:",total_rainfall,"mm")
```

### DAYS WITH THE MOST RAINFALL

```python
rainiest_days=dataset.sort_values(by="Rainfall (mm)", ascending=False).head(5)
print("Top 5 rainiest days:" , rainiest_days)
```

### TEMPERATURE AND RAINFALL CORRELATION

```python
correlation=dataset["Temperature (°C)"].corr(dataset["Rainfall (mm)"])
if correlation > 0:
    print("Hotter days=more rain")
else:
    print("Hotter days=less rain")

    
```

### HUMIDITY AND RAINFALL CORRELATION

```python
HR_correlation=dataset["Humidity (%)"].corr(dataset["Rainfall (mm)"])
if HR_correlation<0:
    print("Less rain=less humid")
else:
    print("More rain=More humid")
```
